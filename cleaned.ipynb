```python
#convert the height in cm to inches
inches_list = []
indexes = []

for i in range(len(fc2_copy['height'])):
    if "'" in fc2_copy.loc[i, 'height']:
        indexes.append(i)
        inches_list.append(fc2_copy.loc[i, 'height'])
```

```python
inches_list
```

```python
#convert height to inches
def convert_height_to_cm(height):
    feet, inches = height.split("'")
    feet = int(feet)
    inches = int(inches)
    height_in_cm = (feet * 30.48) + (inches * 2.54)
    return height_in_cm
```

```python
for i in indexes:
    new_height = convert_height_to_cm(fc2_copy.loc[i, 'height'])
    fc2_copy.loc[i, 'height'] = new_height
```

```python
# changing the height datatpe to float
fc2_copy['height'] = fc2_copy['height'].astype('float')
```

```python
fc2_copy.head()
```

```python
#renaming the height
fc2_copy = fc2_copy.rename(columns={'height': 'height (cm)'})
```
